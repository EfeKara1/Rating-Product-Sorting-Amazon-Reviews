```python
df.helpful_yes=df.helpful_yes.str.replace("["," ")
```

```python
df.total_vote=df.total_vote.str.replace("]"," ")
```

```python
df.total_vote=df.total_vote.astype("int")
```

```python
df.helpful_yes=df.helpful_yes.astype("int")
```

```python
df["helpful_no"]=df.total_vote-df.helpful_yes
```

```python
df.helpful_no.value_counts()
```

```python
def score_pos_neg_diff(pos,neg):
    return pos-neg
```

```python
df["score_pos_neg_diff"]=score_pos_neg_diff(df.helpful_yes.sum(),
                   df.helpful_no.sum())
```

```python
def score_average_rating(pos,neg):
    if pos + neg ==0:
        return 0
    return pos/(pos+neg)
```

```python
df["score_average_rating"]=score_average_rating(df.helpful_yes.sum(),
                   df.helpful_no.sum())
```

```python
import scipy.stats as st
import math
```

```python
def wilson_lower_bound(pos, neg, confidence=0.95):
    """
    Wilson Lower Bound Score hesapla

    - The lower limit of the confidence interval to be calculated for the Bernoulli parameter p is accepted as the WLB score.
    - The score to be calculated is used for product ranking.
    - Note: If the scores are between 1-5, it will be marked as 1-3 down, 4-5 up and made suitable for bernoulli. 

    Parameters
    ----------
    pos: int
        positive comment number
    neg: int
        negative commment number
    confidence: float
        confidence interval

    Returns
    -------
    wilson score: float

    """
    n = pos + neg
    if n == 0:
        return 0
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    phat = 1.0 * pos / n
    return (phat + z * z / (2 * n) - z * math.sqrt((phat * (1 - phat) + z * z / (4 * n)) / n)) / (1 + z * z / n)
```

```python
df["WilsonLB"]=wilson_lower_bound(df.helpful_yes.sum(),df.helpful_no.sum())
```
